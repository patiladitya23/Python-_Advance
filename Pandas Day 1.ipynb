```python
# reading excel file

pd.read_excel('airline1.xls').head()
```

```python
# Incase of multiple files in excelsheet pandas will bydefault read very first sheet always
# But incase we want to read 2nd sheet or third using use procedure given below

pd.read_excel('airline1.xls', sheet_name = "Sheet2")
```

```python
pd.read_excel('airline1.xls', sheet_name = "Sheet3")
```

```python
# Above in sheet3 does not have column name and still showing first row as column name
# but we dont want 1st row as header for that we use below given procedure

pd.read_excel('airline1.xls', sheet_name = "Sheet3", header = None)
```

```python
# Now if want to give our own column name or update this data with our new column name
# For that we use below given procedure

pd.read_excel('airline1.xls', sheet_name = "Sheet3", header = None, names = ['A','B','C','D','E','F'])
```

```python
# No of column name must br equal to No of columns

pd.read_excel('airline1.xls', sheet_name = "Sheet3", header = None, names = ['A','B','C','D'])
```

```python
# Now how to save this data ?
# As of now this file is in notebook

data1 = pd.read_excel('airline1.xls', sheet_name = "Sheet3", header = None, names = ['A','B','C','D','E','F'])
```

```python
data1.to_csv('test1.csv')
```

```python
# If we check test1 file it is showing new column of index
# There is way we can avoid it
data1.to_csv('test2.csv', index = False)
```
