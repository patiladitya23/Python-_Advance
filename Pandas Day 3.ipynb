```python
import pandas as pd
```

```python
# Dictionery
d1 = {
'Key1':[1,2,3,4,5],
'Key2':[1,2,3,4,5,6]
}
```

```python
# Converting dict to pandas dataframe

pd.DataFrame(d1)
# it is showing error coz no of elements in the array must be equals.
```

```python
d2 = {
'Key1':[1,2,3,4,5],
'Key2':[1,2,3,4,5]
}
# No of elements in the arrays of individual keys is same.
```

```python
df1 = pd.DataFrame(d2)
```

```python
# updating index name

pd.DataFrame(d2, index=['a','b','c','d','e'])
```

```python
d3 = {
'Key1':[1,2,3,4,'abc'],
'Key2':[1,2,'xyz',4,5],
    'key3':[1,2,'hello','hi',5]
}
```

```python
df2 = pd.DataFrame(d3)
```

```python
df1
```

```python
df2
```

```python
# Merge Operation

pd.merge(df1,df2)
```

```python
# why we are getting obve result
# it is doing inner  join -> giving common rows of df1 and df2 same as set intersections
```

```python
pd.merge(df1,df2,how='inner',on='Key1')
```

```python

d4 = {
'Key1':[1,2,3,4,5],
'Key2':[1,2,3,4,5]
}
```

```python
d5 = {
'Key1':[1,2,3,4,5],
'Key2':[1,2,'xyz',4,5],
'key3':[1,2,'hello','hi',5]
}
```

```python
df3 = pd.DataFrame(d4)
```

```python
df4 = pd.DataFrame(d5)
```

```python
df3
```

```python
df4
```

```python
pd.merge(df3,df4,how='inner',on='Key1')
```

```python
pd.merge(df3,df4,how='inner',on='Key2')
```

```python
# Left Merge

pd.merge(df3,df4,how='left',on='Key1')
```

```python
# Right Merge

pd.merge(df3,df4,how='right',on='Key2')
```

```python
pd.merge(df3,df4,how='outer',on='Key2')
```

```python
pd.merge(df3,df4,how='outer',on=['Key2','Key1'])
```
