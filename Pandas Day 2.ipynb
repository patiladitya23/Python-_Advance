```python
'''
0-170.77      = A
170.77-341.54 = B
341.54-512.32 = C
'''
```

```python
def df_fare(x):
    if x <= 170.77:
        return "A"
    elif 170.77 < x <=  341.54:
        return "B"
    elif 341.54 < x <= 512.32:
        return "C"
    else:
        return x
```

```python
data['Fare'] = data['Fare'].apply(df_fare)
```
